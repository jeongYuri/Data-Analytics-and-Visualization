**개나리 개화시기 분석하기**

> 꽃의 개화(성장 온도):평균 온도- 기준온도 값의 누적된 합이 96.1도가 되었을때 개화



> 그래프.jpg





생물 계절 모형에 의하면, 꽃눈은 추운 시간이 어느 정도 쌓일때까지 내생휴먼에 들어갑니다. 내생휴면에서 냉각량이 어느 정도 쌓여야 꽃을 피우고, 이 시간이 부족하면 꽃이 피지 않습니다. 개나리의 냉각량은 -90도라고 합니다.
내생 휴면이 끝난 후 바깥 온도가 너무 낮으면 '강제 휴면' 상태에 들어간다고 합니다. 이 시간이 지난 후 온도가 어느 정도 오르면 식물이 꽃을 피우기 위해 일정 시간 따뜻함에 노출 되어야 한다고 합니다. 이 꽃을 피우는데 필요한 온기를 모은 양을 '가온량'이라고 합니다. 가온량은 '일 평균 온도-기준 온도'를 이용하여 구합니다.
개나리의 기준 온도는 4.1도이도이며 평균온도에서 기준온도를 뺀 값을 누적한 값은 84.2도 입니다.



> **개나리의 개화 시기를 분석하기 위하여 "일 평균 온도-기준 온도"를 구한 후 누적값이 84.2도가 되는 날을 구하여 개화시기를 예측 해 볼것입니다.**

내생휴면이 1월에 끝난다는 전제하에 1월부터 꽃의 가온량을 구하여 개화시기를 예측해 보겠습니다.

[2017년,2018년,2019년,2020년,2021년의 1월부터 5월까지 데이터를 이용하였습니다.]





**[2017년도 분석]**

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('mokpoo.csv', 'r', encoding='cp949') 
data = csv.reader(f)
next(data)
temp = []
date = []
summ=0


for row in data :
     row[2]=float(row[2]) #문자열로 이루어진 평균온도를 실수로 변환
     date=row[0]
       
     if 2016 < int(row[0].split('/')[2]) and 2018> int(row[0].split('/')[2]): #2016년~2018년 사이인 2017년 값 구하기
        temp.append(float(row[2])) #평균 기온값 temp에 저장하기
     
        c=np.array(temp)-4.1 #2017년 평균기온 값에서 기준값이 4.1를 뺀 값을 c에 저장
        summ=np.sum(c,axis=0) #temp의 합 구하기
        if summ>85: #가온량 84가 되었을때 멈추기  
          date #해당 날짜 값 얻기
          break        
         
     print(round(summ,2)) #소수점 자리 2번째까지 출력
     print(date)
```

위 값을 통하여 2017년은 4월 11일 개나리가 개화할 것이라 예측할 수 있습니다.
