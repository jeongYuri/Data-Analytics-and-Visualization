- map 함수 적용

```python
# 리스트에 있는 값을 1씩 더해서 새로운 리스트를 만드는 작업 
myList = [1, 2, 3, 4, 5] 

# for 반복문 이용 
result1 = [] 
for i in myList: 
  result1.append(i + 1) 
  
print(result1)
```

- map함수: 순회 가능한 객체의 각 원소에 지정한 함수를 각각 적용
- map(적용할 함수, 순회 가능한 객체)


```python
# map 함수 이용 
def add_one(n): 
  return n + 1
```

```python
myList = [1, 2, 3, 4, 5]
result2 = (map(add_one, myList))
print(result2)

result3 = list(map(add_one, myList)) # map반환을 list 로 변환 
print(result3)
```

```python
a = [1.2, 2.5, 3.7, 4.6]
a = list(map(int,a))
print(a)
```

```python
# 아침 7~9시 승차인원 최대역 찾기
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = 0
mx_station = ''
for row in data :
    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > mx :   # 아침 7~9시 승차인원 인덱스 번호 확인
        mx = sum(row[10:15:2])
        mx_station = row[3] + '(' + row[1] + ')' #역이름(호선)
print(mx_station, mx)
```

### 특정시간의 승차인원이 가장 많은역 찾기

image.png

```python
# 입력으로 받은 특정 시간의 승차인원이 가장 많은 역 찾기
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = 0
mx_station = ''
t = int(input('몇 시의 승차인원이 가장 많은 역이 궁금하세요? : ')) #특정시간 입력받기
for row in data :
    row[4:] = map(int, row[4:])
    a = row[4+(t-4)*2] #특정시간의 승차인원
    if a > mx :
        mx = a
        mx_station = row[3] + '(' + row[1] + ')'
print(mx_station, mx)
```

## 모든 시간대별 최대 승차역 이름 및 승차인원 나타내기 (실습)

```python
# 모든 시간대별(새벽4시~다음날 새벽3시) 최대 승차역 이름 및 승차인원
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24   # 시간대별 최대 승차 인원 저장 리스트 초기화
mx_station = [''] * 24   # 시간대별 최대 승차 인원 역 이름 저장 리스트 초기화
for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        a = row[4+j*2]  # j와 인덱스 번호 사이의 관계식 (0일때 4, 1일때 6, 2일때 8, 3일때 10, ...)
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3]
print(mx_station)
print(mx)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'NanumBarunGothic')
#plt.figure(dpi = 300)
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation =90)
plt.show()
```

```python
# 모든 시간대별(새벽4시~다음날 새벽3시) 최대 승차역 이름 및 승차인원, 그래프에 역이름 + 시간대추가
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24
for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        a = row[4+j*2]
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3]+'('+str(j+4)+')'  # 역이름 + 시간대추가
import matplotlib.pyplot as plt
#plt.figure(dpi = 300)
plt.rc('font', family = 'NanumBarunGothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation = 90)
plt.show()
```

## 모든 시간대별 최대 하차역 이름 및 하차인원(실습)

```python
# 시간대별 최대 하차역 이름 및 하차인원
import csv
import matplotlib.pyplot as plt
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24
for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        b = row[5+j*2]
        if b > mx[j] :
            mx[j] = b
            mx_station[j] = row[3]+'('+str(j+4)+')'
#plt.figure(dpi = 300)
plt.rc('font', family = 'NanumBarunGothic')
plt.bar(range(24), mx, color = 'b')
plt.xticks(range(24), mx_station, rotation = 90)
plt.show()
```

```python
# 모든 지하철역의 시간대별 승하차 인원을 모두 더하여 그래프로 표현
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
s_in = [0] * 24
s_out = [0] * 24
for row in data :
    row[4:] = map(int, row[4:])
    for i in range(24) :
        s_in[i] += row[i*2+4]               #모든 시간대별 모든역의 승차인원 더해서 저장
        s_out[i] += row[i*2+5]               #모든 시간대별 모든역의 하차인원 더해서 저장
        
import matplotlib.pyplot as plt
plt.figure(dpi = 300)
plt.rc('font', family = 'NanumBarunGothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```

### 실습문제
특정 지하철역의 이름을 입력받아 시간대별 승하차 인원을 그래프로 출력되게 하세요.

```python
# 특정역의 시간대별 승하차 인원을 그래프로 표현
import csv
f = open('subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data)
next(data)
s_in = [0] * 24
s_out = [0] * 24

t = str(input('어떤 역이 궁금하세요? : ')) #특정 역 입력받기

for row in data :
    row[4:] = map(int, row[4:])
    if t in row[3]:
       for i in range(24) :
         s_in[i] = row[(i*2)+4]               #모든 시간대별 모든역의 승차인원 더해서 저장
         s_out[i] = row[(i*2)+5]   

import matplotlib.pyplot as plt
plt.figure(dpi = 50)
plt.rc('font', family = 'NanumBarunGothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```
